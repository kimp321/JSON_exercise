```python
# print names of all countries
for child in document_tree.getroot():
    print(child.find('name').text)
```

```python
# print names of all countries and their cities
for element in document_tree.iterfind('country'):
    print('* ' + element.find('name').text + ':'),
    citynames = ''
    for subelement in element.getiterator('city'):
        citynames += subelement.find('name').text + ', '
    print(citynames[:-2])
```

### 3. 10 ethnic groups with the largest overall populations (sum of best/latest estimates over all countries)

```python
# I need to identify the latest reported total populations for countries

# First, create empty dataframe
country_latest_population_df = pd.DataFrame(columns=['Country', 'Year', 'Population'])

# Second, append that dataframe with Country Name, Max Recent Year, and Max Recent Year Population
for country in document_tree.iter('country'):
    population_dictionary = {}
    for population in country.findall('population'):
        population_dictionary[population.get('year')] = population.text 
    if len(population_dictionary) != 0:
        maximum_recent_year = max(population_dictionary.keys())
        maximum_population = population_dictionary[maximum_recent_year]
    df = pd.DataFrame([[country.find('name').text, int(maximum_recent_year), int(maximum_population)]], columns = ['Country', 'Year', 'Population'])
    country_latest_population_df = country_latest_population_df.append(df)

country_latest_population_df.head().set_index('Country')
```

```python
# I need to identify the ethnic group percentage and ethnic group total population for each country

# Create an empty list
EG = []

# Iterate over all countries in doc, identify ethnic group, calculate EG Total Pop, and add data to empty list EG.
for country in document_tree.iterfind('country'):
    #population = int(element.find('population').text)
    for ethnic in country.iterfind('ethnicgroup'):
        data = [country.find('name').text, ethnic.text, float(ethnic.attrib['percentage'])]
        EG.append(data)
        
# Create Data Frame EG_df using list EG
EG_df = pd.DataFrame(EG, columns=['Country', 'Ethnic Group', 'Percentage'])
EG_df.set_index('Country').head(10)
```

```python
# I need to merge the two DFs
merge_DF = country_latest_population_df.merge(EG_df, on='Country', how='left').set_index('Country')
merge_DF
```

```python
# Set pandas so that all floats have 2 decimal places and scientific notation is suppressed
pd.set_option('display.float_format', lambda x: '%.2f' % x)

# Coerce numerical values and convert percentages to decimals
merge_DF['Percentage'] = merge_DF['Percentage'] / 100

# Create new column for the ethnic population for each ethnic group in each country
merge_DF['Ethnic Population'] = merge_DF.Percentage * merge_DF.Population
merge_DF.head()
```

```python
# groupby Ethnic Group, find the .sum for those Ethnic Groups, and sort by ascending population
merge_DF_sum = merge_DF.groupby('Ethnic Group').sum()
merge_DF_sum.sort_values(by='Ethnic Population', ascending=False).head(10)
```

```python
# Delete unwanted columns in merge_df_sum dataframe
Ans = merge_DF_sum.drop('Population', axis = 1)
Ans = Ans.drop('Percentage', axis = 1)
Ans = Ans.drop('Year', axis = 1)
Ans = Ans.sort_values(by='Ethnic Population', ascending=False).head(10)
Ans
```

### 4. Name and country of a) longest river, b) largest lake and c) airport at highest elevation

#### a) longest river

```python
# Set country name and river name to none and the length to 0 
country_=None
river_name=None
length_=0

# Iterate over each element'river' in the document and its length. Create an algorithm so that if the length above
# is less than the length of a given river, it will be replaced with the new, longer river, and the 
# name and country in which the river is located are updated as well.
for river in document_tree.iterfind('river'):
    for length in river.iterfind('length'):
        if length_<float(length.text):
            length_=float(length.text)
            river_name=river.find('name').text
            country_=river.attrib['country']

print('Longest river: ', river_name)
print('Length (KM): ', length_)
print('Countries: ', country_)
```

#### b) largest lake

```python
country = None
lake_name = None
lake_area = 0

for lake in document_tree.iterfind('lake'):
    for area in lake.iterfind('area'):
        if lake_area < float(area.text):
            lake_area = float(area.text)
            lake_name = lake.find('name').text
            country = lake.attrib['country']

print('Largest Lake: ', lake_name)
print('Lake area: ', lake_area)
print('Countries: ', country)
```
